```python
import pandas as pd

# Load Data
churn = pd.read_csv("../data/customer_churn.csv")
sales = pd.read_csv("../data/sales_data.csv")

# Check Missing Values
print(churn.isnull().sum())
print(sales.isnull().sum())

# Remove Duplicates
churn.drop_duplicates(inplace=True)
sales.drop_duplicates(inplace=True)

# Fill Missing Values
churn.fillna(method='ffill', inplace=True)
sales.fillna(method='ffill', inplace=True)

# Save Cleaned Data
churn.to_csv("../data/cleaned_churn.csv", index=False)
sales.to_csv("../data/cleaned_sales.csv", index=False)

print("Data Cleaning Completed")
```
