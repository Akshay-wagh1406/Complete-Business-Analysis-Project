```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Data
churn = pd.read_csv("../data/cleaned_churn.csv")
sales = pd.read_csv("../data/cleaned_sales.csv")

# Churn Distribution
sns.countplot(x='Churn', data=churn)
plt.title("Customer Churn")
plt.show()

# Monthly Charges
sns.histplot(churn['MonthlyCharges'], kde=True)
plt.title("Monthly Charges")
plt.show()

# Sales by Region
sns.barplot(x='Region', y='Sales', data=sales)
plt.title("Sales by Region")
plt.show()

# Profit vs Sales
sns.scatterplot(x='Sales', y='Profit', data=sales)
plt.title("Profit vs Sales")
plt.show()

# Heatmap
sns.heatmap(churn.corr(numeric_only=True), annot=True)
plt.show()
```
